## Overview

In this notebook we use a pre-trained Resnet to label our images.

## A Brief Intro to Resnets

[Resnets](https://paperswithcode.com/method/resnet) are some of the most powerful [convolutional neural networks](https://en.wikipedia.org/wiki/Convolutional_neural_network). Resnets are great for image classification and are also the backbone of networks for other related tasks, such as [image segmentation](https://en.wikipedia.org/wiki/Image_segmentation) and [object detection](https://en.wikipedia.org/wiki/Object_detection). Several varieties of Resnet come pretrained in the `torchvision.models` module. These Resnets have been trained on the [Imagenet dataset](https://www.image-net.org/), which contains 14 million images and a thousand different object classes, which you will find in this notebook.

### Downloading a Pre-trained Resnet34

```python
import torch
import torchvision.transforms as transforms
from torchvision import models
from PIL import Image
from torchvision.models import ResNet34_Weights


# Download pre-trained ResNet-34 model
model = models.resnet34(weights=ResNet34_Weights.IMAGENET1K_V1)
# Setup the model for inference, this makes evaluation faster and consumes less memory
# With model.eval() we let PyTorch know that we are not going to be updating the model's weights
model.eval()
```

```python
# We can check the model's architecture
# Notice that we have 1000 output units in the last layer
model
```

### Define Image Transformations

```python
# We need to make our images use the same means and standard deviations for each channel
# As what we have on the ImageNet dataset
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# Load and preprocess image
image = imag.convert("RGB")
input_image = transform(imag).unsqueeze(0)

```

```python
# How the network 'sees' the image now
import numpy as np
tensor_np = input_image.squeeze().numpy().astype(np.uint8)
# Change layout from (C x H x W) to (H x W x C)
tensor_np = np.transpose(tensor_np, (1, 2, 0))

# Create PIL image
pil_image = Image.fromarray(tensor_np)
pil_image
```

```python
# We inspect the image in its original size (not the thumbnail)
image
```

```python
# Perform inference
with torch.no_grad():
    output = model(input_image)

# Get predicted label
_, predicted = output.max(1)
print(f'Predicted class label: {predicted.item()}')
```

```python
# Position of the index with maximum value
torch.argmax(output)
```

```python
# Shape of the output tensor
output.shape
```

```python
# output.squeeze() allows us to get rid of the last value
# Notice that this is not a probability, but a "logit" (score)
output.squeeze()[torch.argmax(output)]
```

### Mapping Scores to Probabilities

We use the [softmax function](https://en.wikipedia.org/wiki/Softmax_function) to turn the scores that appear in the last layer of the network into probabilities. This makes our output more interpretable.

```python
import torch.nn.functional as F

# Apply softmax to get probabilities
probabilities = F.softmax(output, dim=1)

# torch.argmax(probabilities) returns the position
# Where the tensor max value, which is the label for the image
probabilities.shape, torch.argmax(probabilities), probabilities.squeeze()[torch.argmax(probabilities)]
```

```python
import torch
from torch.utils.data import DataLoader, TensorDataset

# Assuming 'images' is a list of PyTorch tensors of the same shape
images_tensor = torch.stack(images)  # Stack all images into a single tensor

# Create a dataset from tensors
dataset = TensorDataset(images_tensor)

# Create a DataLoader to batch the images
batch_size = 32
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

outputs = []

# Iterate over the DataLoader
for batch in dataloader:
    # Each 'batch' is a tuple with one element (since there are no labels)
    images_batch = batch[0]
    # Now, 'images_batch' is a batch of 32 images you can pass to your model
    # Perform operations with 'images_batch' here
    with torch.no_grad():
      batch_output = model(images_batch)

    #Experiment: try changing this to F.sigmoid(), what is the effect on the output?
    probabilities_batch = F.softmax(batch_output, dim=1)

    for probabilities in probabilities_batch:
      # Get the top 5 labels and their probabilities
      top5_prob, top5_label_idx = torch.topk(probabilities, 5)
      top5_prob = top5_prob.tolist()
      top5_label_idx = top5_label_idx.tolist()
      # Map indices to string labels
      top5_labels = [label_mapping.get(idx, "Unknown") for idx in top5_label_idx]
      top5_output = [{f"label_{index+1}": label, f"probability_{index+1}": f'{prob:.3f}'}
               for index, (label, prob) in enumerate(zip(top5_labels, top5_prob))]

      outputs.append(top5_output)
```
